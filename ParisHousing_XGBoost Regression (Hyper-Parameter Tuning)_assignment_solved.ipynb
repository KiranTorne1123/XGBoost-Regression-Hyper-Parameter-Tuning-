```python
# 1. Histograms for each numeric feature vs price

numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_features.remove('price')

plt.figure(figsize=(15, 20))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(5, 4, i)
    sns.histplot(df, x=feature, y='price', bins=30, cmap='viridis')
    plt.title(f'Histogram of {feature} vs Price')
plt.tight_layout()
plt.show()
```

```python
# From the above we can see direct co-relation of the floor space square metres and the price.
# Rest features hardly have any co-relation with the price.
```

```python
# 2. Scatter plots for numeric features vs price

numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_features.remove('price')
plt.figure(figsize=(15, 20))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(5, 4, i)
    sns.scatterplot(x=df[feature], y=df['price'])
    plt.title(f'{feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
# Similar inference can be drawn for the above scatter plot.
```

```python
# 3. Correlation Heatmap

plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# From above heatmap it is very clear the co-relation of square metres and price.
```

```python
# 4. Box plot for each numeric feature

plt.figure(figsize=(15, 10))
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(df[feature])
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# The above box plots provide information that all the data is within the range and we do not have any outliers.
```

```python
# From all the above charts HEATMAP givees us the best information about the exact perspective of this dataset.
```

```python
# Segregating the Dataset into Input(x) and Output(y)
```

```python
df
```

```python
x = df.iloc[:,:-1].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=17)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Models
```

```python
# We import additional libraries

import xgboost as xgb
from xgboost import XGBRegressor
print('importing is done')
```

```python
xgb = XGBRegressor()
```

```python
# Training the Models

xgb.fit(x_train,y_train)
```

```python
# Predicting the Result Using the Trained Models

y_pred = xgb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import mean_absolute_error,  r2_score
print('importing is done')
```

```python
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
```

```python
# Calculate R-squared score

r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# Results:
# From the above we can see the MAE is 1177541.5404687505 units whereas R2 score is 0.9999 (which is a very good score)
```

```python
# If we keep split as 70/30

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=17)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# There are slight changes in the values once the split ratio is changed.
```

```python
# Select a single test data point from x_test

test_index = 2 
x_test_single = x_test[test_index].reshape(1, -1)

# Getting the true mark for the selected test data point from y_test
true_marks = y_test[test_index]

# Using the trained model to predict the marks for the single test data point
predicted_marks = xgb.predict(x_test_single)

print('Prediction is done')
```
