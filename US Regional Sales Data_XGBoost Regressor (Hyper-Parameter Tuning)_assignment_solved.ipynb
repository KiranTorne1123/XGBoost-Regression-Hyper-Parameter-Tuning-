```python
# 1) Removing commas from numeric columns
```

```python
# We start with Unit Cost column
df['Unit Cost']
```

```python
df['Unit Cost'] = df['Unit Cost'].str.replace(',', '').astype(float)
print ('Replaced Successfully')
```

```python
# The Unit Cost column is changed to float.
df['Unit Cost']
```

```python
# Now we do the same for Unit Price column
df['Unit Price']
```

```python
df['Unit Price'] = df['Unit Price'].str.replace(',', '').astype(float)
print ('Replaced Successfully')
```

```python
# The columns data types has changed to float
df['Unit Price']
```

```python
df.head(5)
```

```python
# 2) Converting the date fields to datetime objects

date_columns = ['ProcuredDate', 'OrderDate', 'ShipDate', 'DeliveryDate']

for column in date_columns:
    df[column] = pd.to_datetime(df[column], format='%d/%m/%y')
print('Conversion completed successfully')
```

```python
# Recalling our dataset

df.head(5)
```

```python
# From the above we can see that the date format has changed
```

```python
# 3) Calculating the order processing times

df['Procured to Order Days'] = (df['OrderDate'] - df['ProcuredDate']).dt.days

df['Order to Ship Days'] = (df['ShipDate'] - df['OrderDate']).dt.days

df['Ship to Delivery Days'] = (df['DeliveryDate'] - df['ShipDate']).dt.days
```

```python
# From the above, we have created 3 more columns namely (Procured to Order Days,Order to Ship Days & Ship to Delivery Days)
# So now we have 19 columns in total.
df.describe()
```

```python
df.dtypes
```

```python
# 4) Encoding the last 3 remaining columns

df = pd.get_dummies(df, columns=['Sales Channel', 'WarehouseCode', 'CurrencyCode'], drop_first=True)
print('Encoding Done Successfully')
```

```python
df.dtypes
```

```python
df
```

```python
# 5) Calculating Total Revenue, Total Cost, and Total Profit

df['Total Revenue'] = df['Order Quantity'] * df['Unit Price'] * (1 - df['Discount Applied'])
df['Total Cost'] = df['Order Quantity'] * df['Unit Cost']
df['Total Profit'] = df['Total Revenue'] - df['Total Cost']
print('Results processed successfully')
```

```python
# 6) Remove non-numeric characters from 'OrderNumber' and convert to float

df['OrderNumber'] = df['OrderNumber'].str.extract('(\d+)').astype(float)
```

```python
# Seasonal Analysis

# Converting date columns to datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])

# Extracting month and year for seasonal analysis
df['Month'] = df['OrderDate'].dt.to_period('M')
# Aggregating data by month
monthly_data = df.groupby('Month').agg({
    'Total Revenue': 'sum',
    'Total Cost': 'sum',
    'Total Profit': 'sum'
}).reset_index()

# Converting 'Month' back to datetime for plotting
monthly_data['Month'] = monthly_data['Month'].dt.to_timestamp()
# Plotting the seasonal data
plt.figure(figsize=(14, 8))

# Plotting Total Revenue
sns.lineplot(data=monthly_data, x='Month', y='Total Revenue', label='Total Revenue', marker='o')

# Plotting Total Cost
sns.lineplot(data=monthly_data, x='Month', y='Total Cost', label='Total Cost', marker='o')

# Plotting Total Profit
sns.lineplot(data=monthly_data, x='Month', y='Total Profit', label='Total Profit', marker='o')

plt.title('Seasonal Analysis of Total Revenue, Total Cost, and Total Profit')
plt.xlabel('Month')
plt.ylabel('Value')
plt.legend()
```

```python
# From the above chart we can see that the sales for month JAN to MAY 2019 have been very low (can be due to off season)
# Rest most of the month we can see there are seasonal buying pattern causing wavy pattern in the line graph.
```

```python
# Select features and target variable

features = ['Order Quantity', 'Discount Applied', 'Unit Cost', 'Procured to Order Days', 'Order to Ship Days', 'Ship to Delivery Days']
target = 'Total Revenue'
```

```python
# We will be using the above as it is for further processing.
```

```python
# Scaling features

from sklearn.preprocessing import StandardScaler
print('Importing is done')
```

```python
scaler = StandardScaler()
scaler
```

```python
df[features] = scaler.fit_transform(df[features])
```

```python
df
```

```python
# Segregating the Dataset into Input(x) and Output(y)

x = df[features].values
x
```

```python
y = df[target].values
y
```

```python
x.shape, y.shape
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('Importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=29)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Models
```

```python
# We import additional libraries

import xgboost as xgb
from xgboost import XGBRegressor
print('Importing is done')
```

```python
xgb = XGBRegressor()
```

```python
# Training the Models

xgb.fit(x_train,y_train)
```

```python
# Predicting the Result Using the Trained Models

y_pred = xgb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import mean_absolute_error,  r2_score
print('importing is done')
```

```python
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
```

```python
# Calculate R-squared score

r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# We get a good result from the above concluding our model is trained well.
```

```python
# If we keep split as 70/30 and random state as 15.

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=15)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# Better results are obtained from 80/20 split.
```
