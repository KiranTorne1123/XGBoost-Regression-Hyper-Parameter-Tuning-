```python
# Segregating the Dataset into Input(x) and Output(y)
```

```python
df
```

```python
x = df.iloc[:,:-1].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
x.shape, y.shape
```

```python
# Splitting the Dataset into Training and Testing Data
```

```python
# we import additional libraries

from sklearn.model_selection import train_test_split
print('importing is done')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=17)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# Loading the Models
```

```python
# We import additional libraries

import xgboost as xgb
from xgboost import XGBRegressor
print('importing is done')
```

```python
xgb = XGBRegressor()
```

```python
# Training the Models

xgb.fit(x_train,y_train)
```

```python
# Predicting the Result Using the Trained Models

y_pred = xgb.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Calculating the Accuracy of the Trained Models
```

```python
# We import additional libraries

from sklearn.metrics import mean_absolute_error,  r2_score
print('importing is done')
```

```python
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
```

```python
# Calculate R-squared score

r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# Results:

# MAE of 123.82 indicates that, on average, the predictions made by our model are approximately 123.82 units away from the true values.
# R-squared value closer to 1 indicates a better fit of the model to the data, our model seems to be performing exceptionally well
```

```python
# If we keep split as 70/30
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,random_state=17)
xgb.fit(x_train, y_train)
y_pred = xgb.predict(x_test)
mae = mean_absolute_error(y_test, y_pred) * 100
print("Mean Absolute Error:", mae)
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)
```

```python
# Predicting the Output of Single Test Data using the Trained Model

y_test
```

```python
# Select a single test data point from x_test

test_index = 3 
x_test_single = x_test[test_index].reshape(1, -1)

# Getting the true mark for the selected test data point from y_test
true_marks = y_test[test_index]

# Using the trained model to predict the marks for the single test data point
predicted_marks = xgb.predict(x_test_single)

print('Prediction is done')
```
